```python
#lab1-1
class Stack:
    def __init__(self):
        self.items = []

    def push(self, element):
        self.items.append(element)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from an empty stack")

    def top(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def __len__(self):
        return len(self.items)

    def is_empty(self):
        return len(self.items) == 0

    def __str__(self):
        return str(self.items)

stack = Stack()
stack.push('T1')
stack.push('T2')
stack.push('T3')

print('stack:', stack)
print('stack.is_empty():', stack.is_empty())
print('stack.length():', len(stack))
print('stack.top():', stack.top())
print('stack.pop():', stack.pop())
print('stack:', stack)
```

```python
# Lab2-3
import random

class Environment:
    def __init__(self):
        self.percept = "Dirty"  # Initial percept for testing

    def get_percept(self):
        return self.percept

    def execute_action(self, action):
        print(f"Executing action: {action}")

class Object:
    def __repr__(self):
        return '<%s>' % getattr(self, '__name__', self.__class__.__name__)

    def is_alive(self):
        return hasattr(self, 'alive') and self.alive

    def display(self, canvas, x, y, width, height):
        pass

class Agent(Object):
    def __init__(self, environment):
        self.environment = environment
        self.alive = True

    def program(self, percept):
        return input('Percept=%s; action? ' % percept)

    def act(self):
        percept = self.environment.get_percept()
        action = self.program(percept)
        self.environment.execute_action(action)

class RandomAgent(Agent):
    def __init__(self, environment, actions):
        super().__init__(environment)
        self.program = lambda percept: random.choice(actions)

def RandomVacuumAgent():
    print("Randomly select one action from the vacuum environment")
    actions = ['Right', 'Left', 'Suck', 'NoOp']
    environment = Environment()
    agent = RandomAgent(environment, actions)

    # Simulate the agent's interaction with the environment
    agent.act()

RandomVacuumAgent()
```

```python
#Lab5
import numpy as np
np.random.seed(42)
x = np.random.randn(15)
y = x + np.random.randn(15)


def jaccard_distance(arr1, arr2):
    intersection = np.intersect1d(arr1, arr2)
    union = np.union1d(arr1, arr2)
    jaccard_dist = 1.0 - len(intersection) / len(union)
    return jaccard_dist

jaccard_dist = jaccard_distance(x, y)
print("Jaccard Distance:", jaccard_dist)


def hamming_distance(arr1, arr2):
    hamming_dist = np.sum(arr1 != arr2) / float(len(arr1))
    return hamming_dist


hamming_dist = hamming_distance(x, y)
print("Hamming Distance:", hamming_dist)


def kendall_tau(list1, list2):
    concordant, discordant = 0, 0
    n = len(list1)

    for i in range(n):
        for j in range(i + 1, n):
            if (list1[i] < list1[j] and list2[i] < list2[j]) or \
               (list1[i] > list1[j] and list2[i] > list2[j]):
                concordant += 1
            elif (list1[i] < list1[j] and list2[i] > list2[j]) or \
                 (list1[i] > list1[j] and list2[i] < list2[j]):
                discordant += 1

    return (concordant - discordant) / (concordant + discordant)

tau = kendall_tau(x, y)
print("Kendall's Tau:", tau)
```

```python
# Lab6
import numpy as np
import matplotlib.pyplot as plt

def kmeans(X, k, max_iters=100):
    # Centroid random ehluuleh
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]

    for _ in range(max_iters):
        # Data buriig oiroltsooh centroid oruulah
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Centroid shinechleh
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

        # Shalgah
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

# Dataset
X, _ = make_classification(n_samples=1000, n_features=2,
                           n_informative=2, n_redundant=0, n_clusters_per_class=1,
                           random_state=4)

# numpy bolgoh
X_np = np.array(X)

# heden shirheg cluster ve
k = 2

# KMeans ehluuleh
labels, centroids = kmeans(X_np, k)

for i in range(k):
    row_ix = np.where(labels == i)
    plt.scatter(X_np[row_ix, 0], X_np[row_ix, 1])

# Graph haruulah
plt.show()
```
