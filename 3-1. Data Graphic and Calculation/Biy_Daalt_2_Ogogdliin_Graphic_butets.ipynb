```python
import pandas as pd
```

```python
file_path = '/content/XAUUSD_D1_eaec3a56-9bf6-4415-936c-b0a08db50bd3.csv'
data = pd.read_csv(file_path, delimiter='\t', index_col=False)

data.head()
```

• 1. high and lows by range -

Range: input (start_date, end_date)

```python
# Convert the 'Time' column to datetime format for filtering
data['Time'] = pd.to_datetime(data['Time']) # 2008-03-13, 2008-03-18

date = input("Enter the start_date and end_date (YYYY-MM-DD, YYYY-MM-DD): ")
start_date, end_date = date.split(",")

# Filter data within the specified date range
filtered_data = data[(data['Time'] >= start_date) & (data['Time'] <= end_date)]

# Find the high and low values within this range
high_value = filtered_data['High'].max()
low_value = filtered_data['Low'].min()

high_value, low_value
```

• 2. Moving average - for each candle

day: input, rate_type: input

```python
day_period = int(input("Enter day: ")) # 2

rate_type = input("Enter rate type: ")# Open
data['Time'] = pd.to_datetime(data['Time'])

date = input("Enter the start_date and end_date (YYYY-MM-DD, YYYY-MM-DD): ") # 2008-03-13, 2008-03-18
start_date, end_date = date.split(",")

filtered_data = data[(data['Time'] >= start_date) & (data['Time'] <= end_date)]

# Adding a new column for the 5-day moving average of 'Open' rates
filtered_data[f'{day_period}_day_MA_{rate_type}'] = data[rate_type].rolling(window=day_period).mean()

# Display the resulting data with the new moving average column
filtered_data[[rate_type, f'{day_period}_day_MA_{rate_type}']].head(10)
```

• 3. Volume Weighted Moving Average - for each candle

(volume * rate) => average - day: input, rate_type: input

```python
# (volume * rate) => average - day

# Define parameters
day = int(input("Enter day: ")) # 3
rate_type = input("Enter rate type: ") # Close
C
data['Time'] = pd.to_datetime(data['Time'])

date = input("Enter the start_date and end_date (YYYY-MM-DD, YYYY-MM-DD): ") # 2008-03-13, 2008-03-18
start_date, end_date = date.split(",")

filtered_data = data[(data['Time'] >= start_date) & (data['Time'] <= end_date)]

# Calculate the VWMA according to the formula
filtered_data[f'{day}_day_{rate_type}'] = (
    data['Volume'] * data[rate_type]
).rolling(window=day).mean() / data['Volume'].rolling(window=day).mean()

# Display the results to verify the calculation
filtered_data[[rate_type, 'Volume', f'{day}_day_{rate_type}']].head(10)
```
