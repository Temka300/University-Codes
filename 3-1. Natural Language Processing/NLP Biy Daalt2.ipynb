```python
import pandas as pd
from collections import Counter
import re
```

```python
# Define a function to preprocess and tokenize text
def preprocess(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text.split()  # Split into words

# Define a function to calculate conditional probabilities with Laplace smoothing
def calculate_probabilities(word_counts, vocab, total_tokens):
    probabilities = {}
    for word in vocab:
        probabilities[word] = (word_counts.get(word, 0) + 1) / (total_tokens + len(vocab))
    return probabilities

# Define a function to calculate sentence probability
def sentence_probability(tokens, class_probabilities):
    prob = 1.0
    for token in tokens:
        prob *= class_probabilities.get(token, 1 / (len(vocabulary) + len(tokens)))  # Smoothing for unseen words
    return prob
```

```python
# Step 1 & 2: Define corpora
corpus_tech = [
    "AI is transforming the tech industry.",
    "Quantum computing is the future.",
    "Advances in machine learning are impressive.",
    "AI models are becoming more powerful.",
    "Blockchain technology is revolutionizing finance.",
    "Tech companies are investing in AI.",
    "The metaverse is the next big thing.",
    "Artificial intelligence outperforms humans in tasks.",
    "Cloud computing is essential for modern tech.",
    "5G networks are expanding globally."
]

corpus_sports = [
    "The soccer team won the championship.",
    "A new record was set in the marathon.",
    "The basketball game went into overtime.",
    "Tennis player dominates in the finals.",
    "The Olympics showcase incredible talent.",
    "Fans celebrate the victory of their team.",
    "The football league is more competitive this year.",
    "Cricket match draws millions of viewers.",
    "Cyclist wins the Tour de France.",
    "Athletes train hard for the competition."
]
```

```python
# Step 3: Select a new sentence
new_sentence = "Technology is growing even better"
new_sentence_tokens = preprocess(new_sentence)
```

```python
# Step 4: Preprocess the corpora and compute probabilities
# Tokenize the corpora
tech_tokens = [word for sentence in corpus_tech for word in preprocess(sentence)]
sports_tokens = [word for sentence in corpus_sports for word in preprocess(sentence)]

# Count word frequencies
tech_word_counts = Counter(tech_tokens)
sports_word_counts = Counter(sports_tokens)

# Create vocabulary
vocabulary = set(tech_tokens + sports_tokens)

# Total tokens in each class
tech_total = sum(tech_word_counts.values())
sports_total = sum(sports_word_counts.values())

# Calculate probabilities
tech_probabilities = calculate_probabilities(tech_word_counts, vocabulary, tech_total)
sports_probabilities = calculate_probabilities(sports_word_counts, vocabulary, sports_total)
```

```python
# Step 5: Calculate conditional probabilities for the sentence
tech_sentence_probability = sentence_probability(new_sentence_tokens, tech_probabilities)
sports_sentence_probability = sentence_probability(new_sentence_tokens, sports_probabilities)

# Normalize probabilities to percentages
total_probability = tech_sentence_probability + sports_sentence_probability
tech_percentage = (tech_sentence_probability / total_probability) * 100
sports_percentage = (sports_sentence_probability / total_probability) * 100
```

```python
# Step 6: Verify classification
actual_topic = "Tech"
predicted_topic = "Tech" if tech_percentage > sports_percentage else "Sports"
classification_correct = actual_topic == predicted_topic

# Summary for display
word_prob_summary = pd.DataFrame({
    "Word": new_sentence_tokens,
    "Tech Probability": [tech_probabilities.get(word, 1 / (tech_total + len(vocabulary))) for word in new_sentence_tokens],
    "Sports Probability": [sports_probabilities.get(word, 1 / (sports_total + len(vocabulary))) for word in new_sentence_tokens]
})

# Display results
print("Word-wise Conditional Probabilities:")
print(word_prob_summary)
print("\nSentence Classification Results:")
print(f"Tech Probability: {tech_percentage:.2f}%")
print(f"Sports Probability: {sports_percentage:.2f}%")

print(f"Predicted Topic: {predicted_topic}")
print(f"{actual_topic} classification Correct? {'Yes' if classification_correct else 'No'}")
```
