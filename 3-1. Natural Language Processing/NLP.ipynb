```python
def split_pdf(input_pdf, start_page, end_page):
    with open(input_pdf, "rb") as file:
        reader = PdfReader(file)

        # Ensure the input range is valid
        total_pages = len(reader.pages)
        if start_page < 1 or end_page > total_pages or start_page > end_page:
            print("Invalid page range. Please check the values.")
            return

        # Create a new writer object to save the page range
        writer = PdfWriter()
        for page_num in range(start_page - 1, end_page):
            writer.add_page(reader.pages[page_num])

        # Save the selected pages to a new PDF
        output_pdf = f"split_pages_{start_page}_to_{end_page}.pdf"
        with open(output_pdf, "wb") as output_file:
            writer.write(output_file)
```

```python
import PyPDF2

def pdf_to_text(input_pdf, output_txt):
    with open(input_pdf, "rb") as file:
        reader = PyPDF2.PdfReader(file)

        # Extract text
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()

        # Save to a text file
        with open(output_txt, "w", encoding="utf-8") as text_file:
            text_file.write(text)
        print(f"PDF converted to text and saved as {output_txt}")

# Convert the PDF to a text file
input_pdf = "g-29.pdf"
output_txt = "output.txt"
pdf_to_text(input_pdf, output_txt)
```

```python
def get_book(url):
  raw = requests.get(url).text
  match = re.search(r"\*\*\* START OF THIS PROJECT GUTENBERG EBOOK .* \*\*\*",raw)
  if match:
      start = match.end()
  else:
      start = 0
  stop = re.search(r"II", raw).start()
  text = raw[start:stop]
  return text
```

```python
def preprocess(sentence):
  return re.sub('[^A-Za-z0-9.]+' , ' ', sentence).lower()
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series, DataFrame
from ipywidgets import FloatProgress
from time import sleep
from IPython.display import display
import re
import pickle
```

```python
url = 'http://www.imdb.com/chart/top?ref_=nv_mv_250_6'
```

```python
result = requests.get(url)
c = result.content
soup = BeautifulSoup(c,"lxml")
```

```python
summary = soup.find('div', {'class': 'article'})
# Create empty lists to append the extracted data.
moviename = []
cast = []
description = []
rating = []
ratingoutof = []
year = []
genre = []
movielength = []
rot_audscore = []
rot_avgrating = []
rot_users = []
```

```python
# Extracting the required data from the HTML soup.
rgx = re.compile(r'[\(\)]')  # Fixed regex to remove parentheses
f = FloatProgress(min=0, max=250)
display(f)

# Attempt to find the table directly
table = summary.find('table') if summary else None
if not table:
    print("No table found in the summary.")
else:
    # Iterate over each row in the table
    for row in table.findAll('tr'):
        for sitem in row.findAll('span', {'class': 'secondaryInfo'}):
            s = sitem.find(text=True)
            year.append(rgx.sub("", s))  # Remove parentheses and add to year list

        for ritem in row.findAll('td', {'class': 'ratingColumn imdbRating'}):
            for iget in ritem.findAll('strong'):
                rating.append(iget.find(text=True))
                ratingoutof.append(iget.get('title').split(' ')[3])  # Fixed index

        for item in row.findAll('td', {'class': 'titleColumn'}):
            for href in item.findAll('a', href=True):
                moviename.append(href.find(text=True))
                rurl = 'https://www.rottentomatoes.com/m/' + href.find(text=True)

                try:
                    rresult = requests.get(rurl)
                    rc = rresult.content
                    rsoup = BeautifulSoup(rc, "lxml")  # Added parser for BeautifulSoup
                    try:
                        rot_audscore.append(rsoup.find('div', {'class': 'meter-value'}).find('span', {'class': 'superPageFontColor'}).text)
                        rot_avgrating.append(rsoup.find('div', {'class': 'audience-info hidden-xs superPageFontColor'}).find_all('div')[1].strip())  # Changed contents access
                        rot_users.append(rsoup.find('div', {'class': 'audience-info hidden-xs superPageFontColor'}).contents[3].contents[2].strip())
                    except AttributeError:
                        rot_audscore.append("")
                        rot_avgrating.append("")
                        rot_users.append("")

                    cast.append(href.get('title'))
                    imdb = "http://www.imdb.com" + href.get('href')

                    try:
                        iresult = requests.get(imdb)
                        ic = iresult.content
                        isoup = BeautifulSoup(ic, "lxml")  # Added parser for BeautifulSoup
                        description.append(isoup.find('div', {'class': 'summary_text'}).find(text=True).strip())
                        genre.append(isoup.find('span', {'class': 'itemprop'}).find(text=True))  # Fixed class name
                        movielength.append(isoup.find('time', {'itemprop': 'duration'}).find(text=True).strip())
                    except requests.exceptions.RequestException:  # Catch general request exceptions
                        description.append("")
                        genre.append("")
                        movielength.append("")
                except requests.exceptions.ConnectionError:
                    rot_audscore.append("")
                    rot_avgrating.append("")
                    rot_users.append("")

        f.value += 1  # Update progress bar
        sleep(0.1)  # Sleep to avoid overwhelming the server
```

```python
# Convert lists to pandas series
moviename = Series(moviename)
cast = Series(cast)
description = Series(description)
rating = Series(rating)
ratingoutof = Series(ratingoutof)
year = Series(year)
genre = Series(genre)
movielength = Series(movielength)
rot_audscore = Series(rot_audscore)
rot_avgrating = Series(rot_avgrating)
rot_users = Series(rot_users)
# Create DataFrame and perform analysis
imdb_df = pd.concat([moviename, year, description, genre, movielength, cast, rating, ratingoutof, rot_audscore, rot_avgrating, rot_users], axis=1)
imdb_df.columns = ['moviename', 'year', 'description', 'genre', 'movielength', 'cast', 'imdb_rating', 'imdb_rating_based_on', 'tomatoes_audience_score', 'tomatoes_rating', 'tomatoes_rating_based_on']
imdb_df['rank'] = imdb_df.index + 1
imdb_df.head(1)  # Display the first row of the DataFrame
```

```python
def text_std(input_text):
  words = input_text.split()
  new_words = []
  for word in words:
    word = re.sub(r'[^\w\s]', '', word)
    if word.lower() in lookup_dict:
      word = lookup_dict[word.lower()]
    new_words.append(word) # This line and the next were incorrectly indented. Corrected to be inside the for loop.
  new_text = " ".join(new_words) # Now correctly indented
  return new_text
```
